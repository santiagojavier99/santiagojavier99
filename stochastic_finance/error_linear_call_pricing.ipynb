```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def error(t,s):
    e=np.abs(1-0.4*s*np.sqrt(t)/(norm.cdf(s*np.sqrt(t)/2)-norm.cdf(-s*np.sqrt(t)/2)))
    return e
    
S_0=100
N_time=100
N_sigma=100
sigma=np.linspace(0.1,0.5,N_sigma)
time=np.linspace(0.2,1,N_time)
e=np.zeros((N_sigma,N_time), dtype='float')
for i in range(N_time):
    for j in range(N_sigma):
        e[i,j]=error(time[i],sigma[j])

# Plot the heatmap
plt.figure(figsize=(10, 8))
plt.imshow(e, extent=[time.min(), time.max(), sigma.min(), sigma.max()],
           aspect='auto', origin='lower', cmap='viridis')
plt.colorbar(label='Error')
plt.xlabel('Time (t)')
plt.ylabel('Sigma ()')
plt.title('Heatmap of Computed Error')
plt.show()
```
